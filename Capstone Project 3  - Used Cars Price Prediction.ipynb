## **Evaluation Metrics**

Pada proyek ini digunakan dua metrik evaluasi: MAPE dan MAE, namun fokus utama adalah MAPE.
- MAPE (Mean Absolute Percentage Error): Rata-rata persentase selisih antara nilai prediksi dan aktual terhadap nilai aktual.
- MAE (Mean Absolute Eror): Rata-rata selisih absolut antara prediksi dan nilai aktual.


Dalam hal ini, nilai variabel target yaitu harga mobil sangat besar dengan rentang yang besar, sehingga MAE sangat cocok untuk melihat kesalahan rata-rata ketika variabel target memiliki rentang yang besar, apabila dibandingkan dengan MSE atau RMSE, MAE merupakan representasi yang lebih langsung dari besarnya kesalahan dan juga lebih robust terhadap outliers.

Alasan fokus ke metrics MAPE untuk prediksi harga mobil bekas, dengan alasan sebagai berikut:
| Aspek                                      | Alasan penggunaan MAPE                                                                                                                                                          |
| ------------------------------------------ | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **Interpretasi**                           | MAPE dalam bentuk persentase, sehingga stakeholder mudah memahami tingkat error meski tanpa satuan harga                                                                        |
| **Scale independent**                      | Tidak dipengaruhi skala harga, tetap adil saat harga bervariasi besar.                                                                                                          |
| **Outlier handling/penanganan outlier**    | Lebih robust/lebih tahan terhadap outliers.                                                                                                                                      |
| **Nilai data target**                      | Cocok untuk data target variabel harga mobil yg nilainya relatif besar.                                                                                                         |

Pada analisis ini meskipun menggunakan MAE sebagai salah satu metrics namun tidak difokuskan ke metrics MAE sebagai metrics utama evaluasi model, dikarenakan:
- MAE tidak proporsional errornya dan tidak scale-independent.
- lebih sulit diinterpretasikan bagi stakeholder non-teknis.
- Cenderung bias ke nilai absolut yang besar.

-----

Alasan Tidak Menggunakan RMSE & MSE :
- RMSE: Sangat sensitif terhadap outlier; satu nilai ekstrem dapat memberikan error drastis.
- MSE: Mengkuadratkan error sehingga memberi penalti berlebihan pada outlier sehingga performa terlihat lebih buruk daripada kondisi sebenarnya.

<br>

> Oleh karena itu, dengan menggunakan MAPE sebagai metrics utama, nilai kesalahan rata-rata suatu model prediksi akan lebih jelas karena telah diinterpretasikan dalam bentuk persentase berdasarkan nilai aktual, lebih tahan terhadap outliers, cocok untuk data target harga mobil yang rentang nilainya relatif besar dan variansinya tinggi.

### **Check Outliers (mild dan extreme)**

```python
# Count mild & extreme outliers
def count_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1

    # Mild outlier thresholds (1.5 * IQR)
    mild_lower = q1 - 1.5 * iqr
    mild_upper = q3 + 1.5 * iqr

    # Extreme outlier thresholds (3 * IQR)
    extreme_lower = q1 - 3 * iqr
    extreme_upper = q3 + 3 * iqr

    # Extreme outliers
    extreme_outliers = data[
        (data[column] < extreme_lower) | (data[column] > extreme_upper)
    ]

    # Mild outliers
    mild_outliers = data[
        ((data[column] < mild_lower) | (data[column] > mild_upper)) &
        ~((data[column] < extreme_lower) | (data[column] > extreme_upper))
    ]

    return {
        "mild_count": len(mild_outliers),
        "extreme_count": len(extreme_outliers),
        "mild_lower": mild_lower,
        "mild_upper": mild_upper,
        "extreme_lower": extreme_lower,
        "extreme_upper": extreme_upper
    }

rows = []

for col in numeric_cols:  
    result = count_outliers(df_model, col)
    rows.append([
        col,
        result['mild_count'],
        result['extreme_count'],
        result['mild_lower'],
        result['mild_upper'],
        result['extreme_lower'],
        result['extreme_upper']
    ])

# Buat dataframe summary
outliers_summary = pd.DataFrame(
    rows,
    columns=[
        'Variable',
        'Mild Outliers Count',
        'Extreme Outliers Count',
        'Mild Lower',
        'Mild Upper',
        'Extreme Lower',
        'Extreme Upper'
    ]
)

# Tampilkan hasil
outliers_summary
```

```python
# check the numbers of mild and extreme outliers after outliers handling with treshold setting

# Count mild & extreme outliers
def count_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1

    # Mild outlier thresholds (1.5 * IQR)
    mild_lower = q1 - 1.5 * iqr
    mild_upper = q3 + 1.5 * iqr

    # Extreme outlier thresholds (3 * IQR)
    extreme_lower = q1 - 3 * iqr
    extreme_upper = q3 + 3 * iqr

    # Extreme outliers
    extreme_outliers = data[
        (data[column] < extreme_lower) | (data[column] > extreme_upper)
    ]

    # Mild outliers
    mild_outliers = data[
        ((data[column] < mild_lower) | (data[column] > mild_upper)) &
        ~((data[column] < extreme_lower) | (data[column] > extreme_upper))
    ]

    return {
        "mild_count": len(mild_outliers),
        "extreme_count": len(extreme_outliers),
        "mild_lower": mild_lower,
        "mild_upper": mild_upper,
        "extreme_lower": extreme_lower,
        "extreme_upper": extreme_upper
    }

rows = []

for col in numeric_cols:
    result = count_outliers(df_model, col)
    rows.append([
        col,
        result['mild_count'],
        result['extreme_count'],
        result['mild_lower'],
        result['mild_upper'],
        result['extreme_lower'],
        result['extreme_upper']
    ])

# Buat dataframe summary
outliers_summary = pd.DataFrame(
    rows,
    columns=[
        'Variable',
        'Mild Outliers Count',
        'Extreme Outliers Count',
        'Mild Lower',
        'Mild Upper',
        'Extreme Lower',
        'Extreme Upper'
    ]
)

# Tampilkan hasil
outliers_summary
```

## **Check Multicolinearity**

Multicollinearity adalah kondisi ketika dua atau lebih variabel independen (variabel x) dalam model regresi saling berkorelasi secara kuat, membawa informasi yang redundan ke model (hubungan linear tinggi satu sama lain), bukan hanya dengan target. Jika dua feature sangat berkorelasi tinggi maka model kesulitan menentukan mana yang sebenarnya berpengaruh. Multicollinearity harus dicek karena membuat model regresi tidak stabil, sulit diinterpretasi, tidak dapat diandalkan dan mudah overfit. 

**Variance Inflation Factor (VIF)** digunakan untuk mengukur seberapa besar sebuah feature dipengaruhi feature lain. VIF > 10 = indikasi kuat bahwa feature tersebut harus dipertimbangkan untuk dihapus/ditangani.

```python
# import library for VIF calculation
from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF calculation
def calculate_vif(df_model):
    vif_list = []
    
    for i in range(df_model.shape[1]):
        vif_value = variance_inflation_factor(df_model.values, i)
        vif_list.append(vif_value)
    
    return pd.DataFrame({
        "Variable": df_model.columns,
        "VIF": vif_list
    })

calculate_vif(df_model[['Year','Engine_Size','Mileage']])
```

Dari hasil perhitungan **Variance Inflation Factor (VIF)** :

- `Year`: ~6,46
- `Engine_Size` : ~5,89
- `Mileage` : ~2,53

Semua nilai VIF < 10, hal ini menunjukkan tidak ada indikasi multikolinearitas yang signifikan. Oleh karena itu, feature-feature tersebut tetap dapat digunakan dalam model tanpa perlu dilakukan penghapusan atau modifikasi lebih lanjut.

### **Preprocessing Scheme**

**Encoding:** agar dapat menggunakan semua feature yang dipilih, maka feature-feature categorical harus diubah menjadi feature numerical, akan digunakan ColumnTransformer untuk dapat mengubah tipe data agar dapat diaplikasikan pada pipeline model. <br>

**1. Binary Encoding**: lebih cocok untuk feature yang memiliki kategori lebih besar daripada 5. <br>
- **Features**: `Type`, `Make`, `Region`.

**2. One Hot Encoding**: lebih cocok untuk feature yang memiliki kategori lebih kecil atau sama dengan 5 dan tidak memiliki tingkatan.
- **Features**: `Gear_Type`, `Origin`

**3. Ordinal Encoding** : lebih cocok untuk feature yang memiliki kategori lebih kecil atau sama dengan 5 dan feature yang memiliki tingkatan.
- **Features** : `Options` -> tingkatan kelengkapan feature mobil (Full : 3, Semi Full: 2, Standard : 1).

**Robust Scaler**: penyamaan satuan/skala dari data dengan menggunakan **median dan IQR**, cocok untuk data yang banyak outliersnya.
- **Features** : `Year`, `Engine_Size`, `Mileage`.

```python
# Total Unique Values in Categorical Features

categoric_features = ['Type', 'Make', 'Gear_Type', 'Origin', 'Region', 'Options']

for col in categoric_features:
    print(f"Unique Values pada feature {col} : {df_model[col].nunique()}")
```

```python
# mapping ordinal encoding

ord = [
    {
        'col': 'Options',
        'mapping': {None: 0, "Standard": 1, "Semi Full": 2, "Full": 3}
    }
]

oe = OrdinalEncoder(cols='Options', mapping=ord)
```

```python
# Preprocessing by ColumnTransformer

preproces = ColumnTransformer([
    ('Binary Encoding', BinaryEncoder(), ['Type', 'Make', 'Region']),
    ('OHE', OneHotEncoder(drop='first'), ['Origin', 'Gear_Type']),
    ('Ordinal Encoding', oe, ['Options']),
    ('Robust Scaling', RobustScaler(), ['Year', 'Engine_Size', 'Mileage']),
], remainder='passthrough')

preproces
```

## **Hyperparameter Tuning**

Hyperparameter Tuning bertujuan untuk mengoptimalkan performa model. Selanjutnya dilakukan Hyperparameter Tuning pada Top 2 Models, yaitu Model XGBoost ([XGBoost](https://xgboost.readthedocs.io/en/stable/parameter.html)) dan Model Random Forest ([Random Forest](https://www.geeksforgeeks.org/machine-learning/random-forest-hyperparameter-tuning-in-python/))

#### Parameter Penting untuk Hyperparameter Tuning:

**A. XGBoost** <br>
**1. learning_rate** : Skala kontribusi tiap pohon baru.
- Nilai terlalu rendah → konvergensi lambat, bisa underfitting jika jumlah pohon rendah.
- Nilai terlalu tinggi (>0.2) → overfitting, model “meledak” dari pola noise.
- Rentang Aman: ~0 – 0.1 (ideal).

**2. n_estimators** : Jumlah boosting rounds / pohon.
- Nilai terlalu rendah (<100) → underfitting (kurang belajar)
- Nilai terlalu tinggi (>600) → overfitting + cost komputasi tinggi
- Rentang Aman: 100 – 600

**3. max_depth** : Kedalaman maksimum setiap pohon.
- Terlalu rendah (<3) → underfitting (model terlalu sederhana)
- Terlalu tinggi (>= 8) → overfitting (model bisa mengikuti noise)
- Rentang: 3 – 8 cukup umum.

**4. min_child_weight**: Minimum data di setiap daun pohon.
- Terlalu rendah (<1) → pohon bisa membagi sangat agresif → risiko overfitting
- Terlalu tinggi (>10) → pohon jadi konservatif → bisa underfit
- Rentang: 1 – 5 ideal.

**5. gamma**: Minimum loss reduction untuk membuat split.
- Tinggi (>0.5) → bisa overfitting
- Rendah (0) bisa underfitting
- Rentang: 0 – 0.3 (tergantung kasus).

**6. subsample** : Proporsi sampel yang digunakan untuk tiap tree.
- Rendah (<0.5) → model bisa underfit karena data terlalu sedikit
- Tinggi (1) → resiko overfitting
- Rentang: 0.6 – 0.9.

**7. colsample_bytree** : Proporsi feature yang dipilih per pohon.
- Terlalu besar (1) → risk overfitting (karena semua feature dipakai)
- Terlalu kecil (<0.5) → bisa underfit (hilang informasi)
- Rentang: 0.6 – 0.9.

**8. reg_lambda** : Mengurangi overfitting dengan mengurangi bobot yang terlalu besar supaya model lebih stabil dan mengurangi variansi (Regularization)
- Terlalu besar (>=20) → Model jadi terlalu sederhana, potensi underfit
- Terlalu rendah (<5) → overfit
- Rentang: 5 - 15

**9. reg_alpha** : Mengurangi overfitting dengan mempercepat model & membuat model lebih sparse → bisa menekan feature kurang penting (Regularization)
- Terlalu besar (>=2.0) → Banyak feature penting malah hilang, performa menjadi drop
- Terlalu rendah (<=0.05) → overfit
- Rentang: 0.05 - 2.0

------------------------------

**B. Random Forest** <br>
**1. n_estimators** : Jumlah banyak pohon.
- Nilai terlalu rendah (<100) → model kurang stabil, bisa underfitting
- NIlai terlalu tinggi (>800) → komputasi tinggi, risiko overfit
- Rentang: 200 – 600.

**2. max_depth** : Seberapa dalam tiap pohon.
- Nilai terlalu tinggi (>25) → overfitting
- Nilai terlalu rendah (<4) → underfitting
- Rentang: 5 – 20.

**3. min_samples_split** : Jumlah minimum sampel untuk melakukan split.
- Rendah (<=3) → overfitting
- Tinggi (>15) → underfitting
- Rentang: 3 – 10.

**4. min_samples_leaf** : Minimum sampel di daun pohon.
- Nilai 1 → bisa overfit
- Terlalu besar (>10) → generalisasi buruk
- Rentang: 1 – 5.

**5. max_features** : Jumlah banyak feature yang dipertimbangkan saat split.
- Terlalu tinggi (0.9-1) → overfit
- Terlalu rendah (0.1-0.3) → kehilangan prediktivitas
- Rentang: sekitar 0.6–0.8 dari semua feature.

| Parameter     |        Penjelasan         | Efek                                                                               |
| ------------- | ------------------------- | ---------------------------------------------------------------------------------- |
| **sqrt**      | √p dari total feature (p)   | Diversity tinggi → lebih robust, mengurangi overfitting                            |
| **log2**      | log₂(p)                   | Sangat cegah overfitting tetapi bisa menurunkan sedikit akurasi                    |

**6. boot_strap** : Metode pengambilan sampel data secara acak dengan pengembalian dari dataset asli untuk membuat subset data yang akan dipakai membangun setiap pohon.

| Parameter           | Efek                                                                        |
| ------------------- | --------------------------------------------------------------------------- |
| **bootstrap=True**  | Sampling dengan pengembalian, pohon beda-beda, stabilitas tinggi            |
| **bootstrap=False** | Semua data dipakai tiap pohon, kurang variasi, potensi overfit lebih tinggi |

| MAPE           | Judgement of Forecast Accuracy |
|----------------|-------------------------------|
| Less than 10%  | Highly accurate               |
| 11% to 20%     | Good forecast                 |
| 21% to 50%     | Reasonable forecast           |
| 51% or more    | Inaccurate forecast           |

Source: Klimberg et al. 2010. Forecasting Performance Measures—What are their practical meaning? [Source](https://www.researchgate.net/publication/235312797_Forecasting_performance_measures_-_What_are_their_practical_meaning)

Menurut Setyo dan Abdillah (2024), MAPE mengukur kesalahan dalam bentuk persentase relatif terhadap nilai aktual. Tolok Ukur "Baik": MAPE diinterpretasikan sebagai persentase kesalahan. Secara umum, MAPE < 10% dianggap sangat baik, antara 10% hingga 20% baik, dan lebih dari 20% dianggap tidak memadai. [Source](https://jpti.journals.id/index.php/jpti/article/view/505)

Menurut Alexandra dan Hartomo (2025), jika Mean Absolute Percentage Error (MAPE) > 20%, maka model dianggap kurang akurat dan perlu dilakukan pemodelan ulang. [Source](https://ejurnal.stmik-budidarma.ac.id/index.php/jurikom/article/view/8639)

Hati et al. (2023), dalam jurnalnya juga menyebutkan bahwa Nilai MAPE antara 10% hingga 20% menunjukkan bahwa prediksi cukup akurat. Nilai MAPE di atas 20% menunjukkan bahwa prediksi menjadi kurang akurat. [Source](https://ejournal.radenintan.ac.id/index.php/IJECS/article/view/20899)

### **Preprocessing Scheme (After Adding New Features)**

**Encoding:** agar dapat menggunakan semua feature yang dipilih, maka feature-feature categorical harus diubah menjadi feature numerical, akan digunakan ColumnTransformer untuk dapat mengubah tipe data agar dapat diaplikasikan pada pipeline model. <br>

**1. Binary Encoding**: lebih cocok untuk feature yang memiliki kategori lebih besar daripada 5. <br>
- **Features**: `Type`, `Make`, `Body_Type`, `Region`.

**2. One Hot Encoding**: lebih cocok untuk feature yang memiliki kategori lebih kecil atau sama dengan 5 dan tidak memiliki tingkatan.
- **Features**: `Gear_Type`, `Origin`

**3. Ordinal Encoding** : lebih cocok untuk feature yang memiliki kategori lebih kecil atau sama dengan 5 dan feature yang memiliki tingkatan.
- **Features** : `Options` -> tingkat kelengkapan feature mobil (Full : 3, Semi Full: 2, Standard : 1) dan `Make_Tier` -> tier brand mobil/tingkat kemewahan mobil (Ultra Luxury : 4, Premium Luxury : 3, Midrange : 2, Entry Level : 1)

**Robust Scaler**: penyamaan satuan/skala dari data dengan menggunakan **median dan IQR**, cocok untuk data yang banyak outliersnya.
- **Features** : `Year`, `Engine_Size`, `Mileage`.

```python
# Total Unique Values in Categorical Features

categoric_features = ['Type', 'Body_Type', 'Region', 'Make', 'Make_Tier', 'Gear_Type', 'Origin', 'Options']

for col in categoric_features:
    print(f"Unique Values pada feature {col} : {df_model[col].nunique()}")
```

```python
# mapping ordinal encoding

ord = [
    {
        'col': 'Options',
        'mapping': {None: 0, "Standard": 1, "Semi Full": 2, "Full": 3}
    },
    {
        'col': 'Make_Tier',
        'mapping': {None: 0, 'Entry Level': 1, 'Midrange': 2, 'Premium Luxury': 3, 'Ultra Luxury' : 4}
    }
]

oe = OrdinalEncoder(cols=['Options', 'Make_Tier'], mapping=ord)
```

- **Prediksi lebih stabil**: 
    - Kalau target meningkat eksponensial, log-transform membantu model menangkap tren yang proporsional.

### **XGBoost Model and The Algorithm Working Mechanism**



Source: Ali, Z.H. and Burhan, A.M. 2023. Hybrid machine learning approach for construction cost estimation: an evaluation of extreme gradient boosting model. Asian Journal of Civil Engineering. [Source](https://www.researchgate.net/journal/Asian-Journal-of-Civil-Engineering-2522-011X?_tp=eyJjb250ZXh0Ijp7ImZpcnN0UGFnZSI6InB1YmxpY2F0aW9uIiwicGFnZSI6InB1YmxpY2F0aW9uIiwicHJldmlvdXNQYWdlIjoiX2RpcmVjdCJ9fQ)

XGBoost is a boosting algorithm that is extremely accurate, and effective for handling large and complex datasets. XGBoost can model complex non-linear relationships, offers many tuning options to optimize performance, and provides easily interpretable feature importance. It is stable, reliable, and widely used in industry. Its combination of efficiency, stability, and high accuracy makes XGBoost one of the best algorithms for building machine learning model.

**XGBoost** works by:
1. Start with the training data <br>
The model begins with the dataset (features + target).

2. Build the first tree (f(L₁)) <br>
A first tree is trained to predict the target values.

3. Calculate errors <br>
After the first tree predicts, the models will check which samples are predicted poorly and compute the gradients (how wrong) and Hessians (how confident the model is).

4. Increase the weight of high-error samples <br>
Points with larger errors get more influence (heavier weight). These points become “more important” for the next tree.

5. Build the second tree (f(L₂)) and focusing on high-error points <br>
The new tree tries to fix the remaining errors from the previous tree.
Because weights changed, the tree naturally splits differently.

6. Update predictions again <br>
Combine Tree 1 + Tree 2 predictions (with a learning rate).

7. Repeat the process for many trees (f(Lₙ)) <br>
Each new tree is built on the updated residuals and updated weights. The red arrows in the diagram represent this boosting loop.

8. In testing, each tree makes a prediction <br>
The model applies all trained trees to the new data.

9. Final prediction = average/weighted sum of all trees <br>
All tree outputs are aggregated to produce the final prediction.

**Conclusion of How XGBoost Works**:
1. Train the first tree on the training data.
2. Measure errors and increase weight for samples with large errors.
3. Train the next tree that focuses more on these samples with large errors.
4. Add the new tree’s prediction to the previous prediction.
5. Repeat the process for many trees.
6. Final prediction is the combined output of all trees.

Ketika tingkat error kecil, penetapan harga menjadi lebih akurat sehingga kendaraan cenderung lebih cepat terjual sambil tetap menjaga revenue growth.

Menurut Lewis (1982) :
| MAPE           | Judgement of Forecast Accuracy |
|----------------|-------------------------------|
| Less than 10%  | Highly accurate               |
| 11% to 20%     | Good forecast                 |
| 21% to 50%     | Reasonable forecast           |
| 51% or more    | Inaccurate forecast           |

Dari tabel di atas, nilai **MAPE sebesar ~19%** menjadikan akurasi performa model **XGBoost** berada dalam kategori **Good Forecasting** (Klimberg et al., 2010).

Source: Klimberg et al. 2010. Forecasting Performance Measures—What are their practical meaning? [Source](https://www.researchgate.net/publication/235312797_Forecasting_performance_measures_-_What_are_their_practical_meaning)

```python
# Residual Calculation for Business Impact Analysis

model_xgb = Pipeline([
        ('preprocessing', preproces),
        ('model', xgb_tuned)])

model_xgb.fit(xtrain,ytrain)
y_pred = model_xgb.predict(xtest)

residual_xgb = pd.DataFrame({
    'y test' : ytest,
    'y pred' : y_pred
})

residual_xgb['Residual'] = ytest - y_pred
residual_xgb
```

- Residual > 0 : Underpredict
- Residual < 0 : Overpredict

## **Business Impact Analysis (Cost Benefit Analysis)**

1. Underpredict : harga estimasi lebih rendah dari harga pasar.
- Seller: Potensi rugi atau menolak harga → risiko seller churn.
- Platform: Tidak menimbulkan kerugian finansial langsung.
- Jika hanya memberi rekomendasi harga → transaksi cepat → fee tetap masuk.
- Jika Syarah membeli mobil → margin bisa lebih besar, tetapi supply bisa berkurang jika seller menolak harga.
- Dampak utama: risiko jangka panjang pada supply dan kerugian keuangan.

2. Overpredict : harga estimasi lebih tinggi dari harga pasar. 
- Ini adalah risiko yang paling merugikan perusahaan.
- Mobil dibeli terlalu mahal → dijual kembali susah → rugi secara langsung.
- Harga listing tampak mahal → buyer tidak tertarik → conversion turun.
- Holding/operational cost naik (penyimpanan, depresiasi, logistik, iklan).
- Buyer churn → seller ikut pindah → ekosistem melemah.
- Dampak utama: **kerugian finansial langsung**, penurunan traffic dan risiko buyer churn.

| Dampak                      | Underpredict                                | Overpredict                                           |
| --------------------------- | ------------------------------------------- | ----------------------------------------------------- |
| **Untung/Rugi Syarah**      | Margin kecil / seller menolak → lost supply | **Kerugian langsung (beli kemahalan + holding cost)** |
| **Kecepatan menjual mobil** | Cepat terjual                               | Lama terjual → rugi                                   |
| **Risiko bisnis**           | Moderat                                     | **Tinggi**              |
| **Dampak ekosistem**        | Seller churn                                | Buyer churn + inventory stagnan                       |

Bagi Marketplace, menjaga customer = menjaga ekosistem. Jika customer/buyer meninggalkan platform, maka seller juga akan ikut meninggalkan platform dan bisnis bisa collaps.


```python
print('COST–BENEFIT ANALYSIS')

# 1. Data
n_data = df_model.shape[0]
print(f'Total Data: {n_data:,}')

avg_true_price = np.mean(ytest)
print(f'Average True Price: {avg_true_price:,.2f} SAR')

model_mae = 11832
print(f'Model MAE: {model_mae:,.2f} SAR')

# 2. Underpredict / Overpredict
underpredict = residual_xgb[residual_xgb["Residual"] > 0]  # true > predict
overpredict  = residual_xgb[residual_xgb["Residual"] < 0]  # true < predict

print(f'Underpredict cases: {len(underpredict)}')
print(f'Overpredict cases:  {len(overpredict)}')

prop_under = len(underpredict) / n_data
prop_over  = len(overpredict) / n_data

print(f'Underpredict proportion: {prop_under:.2%}')
print(f'Overpredict proportion:  {prop_over:.2%}')

# 3. Platform Fee Assumptions
service_fee_pct = 0.08     # 8% service fee
listing_fee = 500          # per listing revenue
holding_cost = 195         # SAR/day
days_per_1pct_error = 1.5  # extra days per 1% overpricing

print('\nPLATFORM FEE ASSUMPTIONS')
print(f'Service Fee Percentage: {service_fee_pct*100}%')
print(f'Listing Fee per Listing: {listing_fee} SAR')
print(f'Operational Holding Cost: {holding_cost} SAR/day')

# 4. Baseline Loss WITHOUT MODEL (assumption for manual pricing error 20%)
baseline_error_pct = 0.2   # 20% pricing error
baseline_extra_days = baseline_error_pct * 100 * days_per_1pct_error

baseline_lost_transactions = baseline_error_pct * n_data

baseline_lost_service_fee = baseline_lost_transactions * (avg_true_price * service_fee_pct)
baseline_holding_loss = baseline_lost_transactions * baseline_extra_days * holding_cost

baseline_listing_revenue = listing_fee * n_data

baseline_under_loss = baseline_lost_transactions * (avg_true_price * 0.5 * baseline_error_pct)  # approximate lost margin for underpredict
baseline_over_loss  = baseline_lost_service_fee + baseline_holding_loss

total_baseline_loss = baseline_under_loss + baseline_over_loss

print('\n--------------- BASELINE LOSS / MANUAL (WITHOUT MODEL) ---------------')
print(f'Lost Platform Revenue from Underpredict (Potential Margin Loss): {baseline_under_loss:,.2f} SAR')
print(f'Lost Platform Revenue from Overpredict (Service Fee + Holding Cost): {baseline_over_loss:,.2f} SAR')
print(f'TOTAL LOSS WITHOUT MODEL: {total_baseline_loss:,.2f} SAR')
print(f'Listing Fee Revenue (Unaffected): {baseline_listing_revenue:,.2f} SAR')

# 5. MODEL LOSS (WITH MODEL)
under_lost_margin = (underpredict["Residual"]) * 0.5  # simulate 50% opportunity loss
total_under_loss = under_lost_margin.sum()

over_lost_service_fee = abs(overpredict["Residual"]) * service_fee_pct
over_extra_days = (abs(overpredict["Residual"]) / avg_true_price * 100) * days_per_1pct_error
over_holding_loss = over_extra_days * holding_cost

total_over_loss = over_lost_service_fee.sum() + over_holding_loss.sum()

model_listing_revenue = listing_fee * n_data
total_model_loss = total_under_loss + total_over_loss

print('\n--------------- MODEL LOSS (WITH MODEL) ----------------------')
print(f'Lost Platform Revenue From Underpredict (Potential Margin Loss): {total_under_loss:,.2f} SAR')
print(f'Lost Platform Revenue From Overpredict (Service Fee + Holding Cost): {total_over_loss:,.2f} SAR')
print(f'TOTAL LOSS WITH MODEL: {total_model_loss:,.2f} SAR')
print(f'Listing Fee Revenue (Unaffected): {model_listing_revenue:,.2f} SAR')

# 6. Economic Savings Using Model
total_savings = total_baseline_loss - total_model_loss
savings_pct = total_savings / total_baseline_loss * 100

print('\n--------------------- ECONOMIC SAVINGS -----------------------')
print(f'Revenue Loss Without Model: {total_baseline_loss:,.2f} SAR')
print(f'Revenue Loss Using Model:   {total_model_loss:,.2f} SAR')
print('---------------------------------------------------------------')
print(f'TOTAL SAVINGS USING MODEL: {total_savings:,.2f} SAR')
```

### Insight dari Cost Benefit Analysis :

- Model price prediction berhasil **mengurangi kerugian ekonomi perusahaan sekitar 10,5 juta SAR** dibandingkan tanpa menggunakan model.
- Meskipun terdapat error MAE ~11832 SAR dan MAPE ~19%, model tetap memberikan keuntungan finansial signifikan bagi perusahaan.

Dampak positif penggunaan model ML terhadap bisnis:
- Overpredict dan underpredict berkurang
→ Konversi listing meningkat, risiko seller/buyer churn menurun.

- Mobil lebih cepat terjual
→ Holding/operational cost per unit turun drastis, sehingga biaya operasional perusahaan lebih rendah.

- Harga lebih tepat → kepercayaan buyer & seller naik
→ Traffic meningkat → berdampak positif pada revenue growth jangka panjang.
